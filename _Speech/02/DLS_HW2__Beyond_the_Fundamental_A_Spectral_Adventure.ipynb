## **Introduction**

**Что делает скрипку похожей на скрипку?**

Вы когда-нибудь задумывались, почему скрипка, флейта и гитара звучат совершенно по-разному, даже играя одну и ту же ноту на одинаковой громкости?

Ответ кроется в фундаментальном свойстве звука, называемом тембром. Тембр — это «цвет» или «текстура» звука — то, что позволяет вам мгновенно отличить трубу от фортепиано или голос вашего друга от голоса незнакомца, даже если они поют в одной тональности.

**Но что физически создаёт тембр?**

Всё сводится к гармоникам — смеси частот, сопровождающих основной тон, — и тому, как их амплитуды и фазы изменяются с течением времени. Два инструмента, играющие ноту Ля или А4 (A4 — это нота Ля первой октавы с частотой 440 Гц), оба воспроизводят эту основную частоту, но также генерируют разные наборы обертонов или гармоник (2×440 Гц, 3×440 Гц и т. д.) с уникальной интенсивностью и паттернами. Этот спектральный «отпечаток» называют тембром.

### **1.2 Compute the Fourier Transform and plot the magnitude spectrum** [1 score]
- Compute the FFT of the entire audio signal. Use `np.fft.rfft` and `np.fft.rfftfreq`.
- Compute absolute values of the magnitude with `np.abs()` and normalize it.
- Plot the magnitude spectrum (only up to Nyquist frequency).
- Label axes: frequency (Hz) vs magnitude.

> 💡 **Tip**: Normalize magnitude by dividing by `len(signal)`.

````python
# Compute FFT
# Вычисление FFT
n = len(y)  # Длина аудиофайла
# FFT: Применяем np.fft.rfft для расчета БПФ, так как у нас реальные данные (монофонический сигнал).
fft_result = np.fft.rfft(y)  # БПФ (FFT) для реальных данных
frequencies = np.fft.rfftfreq(n, d=1/sr)  # Частоты, соответствующие БПФ

# Вычисление величины спектра (модуль FFT)
# Нормализация: Магнитуда делится на длину сигнала для нормализации.
magnitude = np.abs(fft_result) / n  # Нормализация

# Проверим максимальную амплитуду спектра
print(f"Max magnitude in spectrum: {np.max(magnitude):.4f}")

# Plot
# Построение графика
# Визуализируем спектр частот в пределах Нюквиствовой частоты (половина sample rate).
plt.figure(figsize=(10, 6))
plt.plot(frequencies, magnitude)
plt.title('Magnitude Spectrum of the Audio Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.xlim(0, sr/2)  # Ограничиваем ось X до Нюквиствовой частоты
plt.grid(True)
plt.show()
````

Для того чтобы ответить на вопросы, мы должны создать функцию ```get_harmonic_amplitude```, которая будет находить амплитуды спектра на частотах гармоник. 

Для этого используем ближайшие пики на спектре.

1. Основная частота (f₀) — это частота фундамента, которая для ноты A4 равна 440 Hz.

2. Мы также вычислим амплитуды на 2-й, 3-й и 5-й гармониках, а затем выведем числовые значения амплитуд для каждой из них.

````python
# Find peaks near harmonic frequencies
# Функция для нахождения амплитуды гармоники в диапазоне ±5% от теоретической частоты
def get_harmonic_amplitude(freq, spectrum, freqs, tolerance=0.05):
    # Ищем индексы частот в пределах 5% от заданной частоты
    idx = np.where(np.abs(freqs - freq) <= tolerance * freq)[0]
    # idx = np.argmin(np.abs(freqs - freq))  # Индекс ближайшей частоты
    if len(idx) > 0:
        return np.max(spectrum[idx])
    else:
        return 0.0

# Получаем амплитуды для фундамента и гармоник
freqs = frequencies
fundamental_amp = get_harmonic_amplitude(f0, magnitude, freqs)
second_harm_amp = get_harmonic_amplitude(2*f0, magnitude, freqs)
third_harm_amp = get_harmonic_amplitude(3*f0, magnitude, freqs)
fifth_harm_amp = get_harmonic_amplitude(5*f0, magnitude, freqs)

# Вывод результатов
print(f"Fundamental frequency (f0): {f0:.1f} Hz")
print(f"Fundamental amplitude: {fundamental_amp:.4f}")
print(f"2nd harmonic amplitude: {second_harm_amp:.4f}")
print(f"3rd harmonic amplitude: {third_harm_amp:.4f}")
print(f"5th harmonic amplitude: {fifth_harm_amp:.4f}")
````

````python
# Построение графика спектра с гармониками
plt.figure(figsize=(10, 6))
plt.plot(frequencies, magnitude)
plt.title('Magnitude Spectrum with Harmonics')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.xlim(0, sr/2)  # Ограничиваем ось X до Нюквиствовой частоты
plt.grid(True)

# Добавление вертикальных линий для гармоник
harmonics = [f0, 2*f0, 3*f0, 5*f0]
for harmonic in harmonics:
    plt.axvline(harmonic, color='r', linestyle='--')
    plt.text(harmonic, max(magnitude)*0.1, f'Harmonic ({harmonic:.1f} Hz)', rotation=90, color='red', ha='center')

plt.show()
````

### **1.6 Count significant harmonics** [1 score]
- Define a "significant harmonic" as one with amplitude > 10% of the max amplitude.
- For **Flute** and **Violin** (same pitch), count how many significant harmonics each has.
- Which instrument is "brighter"? Why?

````python
def get_harmonic_amplitude(freq, spectrum, freqs, tolerance=0.02):
    idx = np.where(np.abs(freqs - freq) <= tolerance * freq)[0]
    if len(idx) > 0:
        return np.max(spectrum[idx])
    return 0.0

def count_significant_harmonics(magnitude, freqs, f0, max_freq=5000, threshold_ratio=0.1):
    max_amp = np.max(magnitude)
    threshold = threshold_ratio * max_amp

    harmonics = []
    n = 1

    while n * f0 <= max_freq:
        freq = n * f0
        amp = get_harmonic_amplitude(freq, magnitude, freqs)

        if amp > threshold:
            harmonics.append((n, freq, amp))

        n += 1

    return len(harmonics), harmonics

# Compare Flute and violin
for inst in ['Flute', 'Violin']:
    path_to_wav = "content\\" + database[inst]
    y_inst, sr_inst = librosa.load(path_to_wav, sr=None)
    # Compute FFT
    Y_inst = np.fft.rfft(y_inst)
    # Get frequencies
    freqs_inst = np.fft.rfftfreq(len(y_inst), d=1/sr_inst)
    # Extract magnitude and normalize it
    magnitude_inst = np.abs(Y_inst) / len(y_inst)

    count, harm_list = count_significant_harmonics(magnitude_inst, freqs_inst, f0)
    print(f"{inst.capitalize()}: {count} significant harmonics")
````

### **1.7. Simple Instrument Classification** [2 score]
- Use dataset `TinySOL_short` provided below with 3 instruments of the **same pitch** A5.
- **Approach A**: Use raw audio waveforms (flattened) as features → train a Random Forest.
- **Approach B**: Use magnitude spectrum (first 1024 bins) as features → train same model.
- Compare accuracy on a test set.
- **Question**: Why does Approach B perform better?

> 🛠️ Use `sklearn.tree.RandomForestClassifier` and `train_test_split`.

First, download dataset.

````python
# download dataset
# !gdown --folder 1wMa4fDtWRtO-YAz-gY5Jdrl94GQZUwmO
````

````python
df = pd.read_csv("content\\TinySOL_short\\TinySOL_short.csv")
df.info()
````

````python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Prepare data for 3 instruments, same pitch (A5)
instruments_class = ['Violin', 'Flute', 'Cello']
# Мы будем использовать 3 инструмента на одной ноте (A5) из TinySOL_short. Подходы: сырые волны и мagnitude spectrum.
pitch_class = 'A5'
df_filtered = df[(df['Instrument (in full)'].isin(instruments_class)) & (df['Pitch'] == pitch_class)]
X_wave = []
X_spec = []
y_labels = []

for idx, row in df_filtered.iterrows():
    audio_path = "content\\" + row['Path']  # assuming CSV has column with audio file paths
    y, sr = librosa.load(audio_path, sr=None)  # load audio
    
    # Approach A: raw waveform (flattened)
    X_wave.append(y[:4096] if len(y) >= 4096 else np.pad(y, (0, 4096 - len(y))))  # fix length
    
    # Approach B: magnitude spectrum (first 1024 bins)
    D = np.abs(np.fft.fft(y))[:1024]
    X_spec.append(D)
    
    y_labels.append(row['Instrument (in full)'])

# Split data
Xw_train, Xw_test, yw_train, yw_test = train_test_split(X_wave, y_labels, test_size=0.3, random_state=42, stratify=y_labels)
Xs_train, Xs_test, ys_train, ys_test = train_test_split(X_spec, y_labels, test_size=0.3, random_state=42, stratify=y_labels)

# Train classifiers
clf_wave = RandomForestClassifier(n_estimators=100, random_state=42)
clf_wave.fit(Xw_train, yw_train)

clf_spec = RandomForestClassifier(n_estimators=100, random_state=42)
clf_spec.fit(Xs_train, ys_train)

# Evaluate
acc_wave = accuracy_score(yw_test, clf_wave.predict(Xw_test))
acc_spec = accuracy_score(ys_test, clf_spec.predict(Xs_test))

print(f"Accuracy with raw waveforms: {acc_wave:.3f}")
print(f"Accuracy with spectra: {acc_spec:.3f}")
````

**Answer**: Approach B (magnitude spectrum) выполняется лучше, чем Approach A (сырые волны), по следующим причинам:

* Тембровая информация:

    Спектр частот (magnitude spectrum) содержит информацию о тембре инструмента, который является основным признаком для различия между инструментами. 

    Даже если два инструмента играют одинаковую ноту (A5), их спектры будут отличаться из-за разных гармоник и формы волны. 

    Таким образом, спектр предоставляет более четкую информацию о характерных признаках звука каждого инструмента, что облегчает классификацию.

* Устойчивость к шуму: 

    Сырые волны могут быть чувствительны к фазовым сдвигам, временным вариациям и другим шумам в аудио-сигнале, что затрудняет извлечение полезных признаков. 

    Спектр же абстрагируется от этих нюансов, фокусируясь на амплитудах частотных компонентов, что делает его менее чувствительным к небольшим изменениям и более стабильным для классификации.

* Снижение размерности: 

    В Approach A используются все данные из волны (все сэмплы), что приводит к более сложной задаче для модели. 

    В Approach B используется только первая половина спектра (первые 1024 бина), что сокращает размерность и снижает риск переобучения, помогая классификатору обучиться на более компактных и информативных признаках.

* Инвариантность к временным смещениям: 

    Спектр менее подвержен смещениям и вариациям во времени. 

    Он отображает частотные компоненты, а не точную временную форму сигнала. 

    В то время как форма волны может изменяться с течением времени, спектр остаётся стабильным для каждого инструмента, что облегчает задачу классификации.

Итог:

    Magnitude spectrum извлекает более структурированную и стабильную информацию о звуке, чем сырые волны, что делает его более подходящим для классификации инструментов, даже если они играют одинаковую ноту.

### **3.3 Create Mel filterbank manually** [2 score]
- Number of Mel bands: `n_mels = 128`.
- Frequency range: 0 to `sr/2`.
- Steps:
  1. Convert Hz to Mel: `mel = 2595 * np.log10(1 + f / 700)`
  2. Create `n_mels + 2` equally spaced points in Mel scale.
  3. Convert back to Hz.
  4. Build triangular filters (each filter overlaps with neighbors).
- Output: a matrix `mel_basis` of shape `(n_mels, n_fft//2 + 1)`.

> 📚 Reference: [Librosa mel filterbank docs](https://librosa.org/doc/main/generated/librosa.filters.mel.html)

````python
def hz_to_mel(frequencies):
    """Convert Hz to Mel scale"""
    return 2595 * np.log10(1 + frequencies / 700)

def mel_to_hz(mels):
    """Convert Mel to Hz scale"""
    return 700 * (10**(mels / 2595) - 1)

def create_mel_filterbank(sr, n_fft, n_mels=128, fmin=0.0, fmax=None):
    """Create Mel filterbank manually"""
    
    if fmax is None:
        fmax = sr / 2
    
    # Frequency bins
    n_freqs = n_fft // 2 + 1
    linear_freqs = np.linspace(0, sr / 2, n_freqs)

    # Mel points
    min_mel = hz_to_mel(fmin)
    max_mel = hz_to_mel(fmax)
    mel_points = np.linspace(min_mel, max_mel, n_mels + 2)
    hz_points = mel_to_hz(mel_points)

    # Create filterbank
    filterbank = np.zeros((n_mels, n_freqs))

    for m in range(n_mels):
        # Triangle vertices
        left = hz_points[m]
        center = hz_points[m + 1]
        right = hz_points[m + 2]

        for k, freq in enumerate(linear_freqs):
            if left <= freq <= center:
                filterbank[m, k] = (freq - left) / (center - left)
            elif center < freq <= right:
                filterbank[m, k] = (right - freq) / (right - center)
            else:
                filterbank[m, k] = 0

    return filterbank

# параметры (должны совпадать с предыдущим шагом)
n_fft = 2048

# Create filterbank
mel_basis_manual = create_mel_filterbank(sr, n_fft, n_mels=128)

print(mel_basis_manual.shape)  # (128, 1025)
````

### **3.4 Apply filterbank to power spectrogram** [0.5 score]
- Compute: `mel_spec_manual = np.dot(mel_basis, S)`

````python
# Apply manual filterbank
mel_spec_manual = np.dot(mel_basis_manual, S)

# проверка формы
print(mel_spec_manual.shape)
````

📌 Пояснение:

* mel_basis_manual имеет форму (128, n_fft//2 + 1)
* S имеет форму (n_fft//2 + 1, time_frames)
* результат mel_spec_manual будет (128, time_frames) — Mel-спектрограмма
* следующим шагом обычно делают логарифмирование (log-Mel spectrogram).

#### 3.7.1 Slaney-нормализация

В librosa фильтры нормализуются по площади, а не просто треугольники.

Добавим это в конец цикла создания фильтров:
```

enorm = 2.0 / (hz_points[2:n_mels+2] - hz_points[:n_mels])
filterbank *= enorm[:, np.newaxis]
```

#### 3.7.2 Построение фильтров через индексы (а не частоты)

librosa работает через биновые индексы, а не напрямую Hz. Это уменьшает расхождения.

Заменим часть с треугольниками

````python
def create_mel_filterbank(sr, n_fft, n_mels=128, fmin=0.0, fmax=None):
    if fmax is None:
        fmax = sr / 2

    # Бины
    n_freqs = n_fft // 2 + 1

    # Mel точки
    min_mel = hz_to_mel(fmin)
    max_mel = hz_to_mel(fmax)
    mel_points = np.linspace(min_mel, max_mel, n_mels + 2)
    hz_points = mel_to_hz(mel_points)

    # Перевод Hz → ближайшие FFT бины
    bin_indices = np.floor((n_fft + 1) * hz_points / sr).astype(int)

    filterbank = np.zeros((n_mels, n_freqs))

    for m in range(n_mels):
        left = bin_indices[m]
        center = bin_indices[m + 1]
        right = bin_indices[m + 2]

        # левая сторона
        for k in range(left, center):
            filterbank[m, k] = (k - left) / (center - left)

        # правая сторона
        for k in range(center, right):
            filterbank[m, k] = (right - k) / (right - center)

    # ✅ Slaney normalization (КЛЮЧ!)
    enorm = 2.0 / (hz_points[2:n_mels+2] - hz_points[:n_mels])
    filterbank *= enorm[:, np.newaxis]

    return filterbank
````

````python
# После этого сравним
mel_basis_manual = create_mel_filterbank(sr, n_fft, n_mels=128)
mel_spec_manual = np.dot(mel_basis_manual, S)

are_close = np.allclose(mel_spec_manual, mel_spec_librosa, atol=1e-5)
print(are_close)
````

* визуально → почти не отличить
* численно → расхождения минимальны
